```python
import streamlit as st
import pickle
import numpy as np

model = pickle.load(open("model.pkl", "rb"))
le = pickle.load(open("encoder.pkl", "rb"))

st.title("GitHub Repository Popularity Predictor")

forks = st.number_input("Forks", 0)
watchers = st.number_input("Watchers", 0)
open_issues = st.number_input("Open Issues", 0)
repo_age = st.number_input("Repo Age (days)", 0)
language = st.selectbox("Language", le.classes_)
is_fork = st.selectbox("Is Fork", [0,1])
has_wiki = st.selectbox("Has Wiki", [0,1])

if st.button("Predict"):
    lang_encoded = le.transform([language])[0]
    input_data = np.array([[forks, watchers, open_issues, repo_age, lang_encoded, is_fork, has_wiki]])
    prediction = model.predict(input_data)[0]
    prob = model.predict_proba(input_data)[0][1]

    st.success(f"Prediction: {'Popular' if prediction==1 else 'Not Popular'}")
    st.write(f"Probability: {prob:.2f}")
```
